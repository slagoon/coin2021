### None
- 아무런 값을 갖지 않을 때 사용
- 일반적으로 변수가 초기값을 갖지 않게 하여 해당 변수를 생성할 때 사용
- 기타 언어의 NULL, nil 등과 같은 의미로 사용

- Python은 변수를 생성할 때 항상 값을 초기화 시켜야 함
- 어떤 값이 들어갈 지 모를 때 사용하는것이 None

```python
c = None
print(c)
```

### comparison operator(비교 연산자)
- 프로그래밍에서 비교를 할 경우, = 대신 == 를 사용
-  <, >
- <=, =>
- ==
- !=
- 비교 연산자의 결과는 bool 타입

```python
a = 5
b = 3
```

```python
print(a > b)
print(a < b)
print(a == b)
print(a != b)

c = a > b
print(c)
print(type(c))
```

### numbers(숫자형 타입)
- 정수, 실수로 구성
- 수학의 기본 연산자(가감승제)사용 가능

```python
a = 10
b = 5

print(a + b) # 더하기
print(a - b) # 빼기
print(a * b) # 곱하기
print(a % b) # 나머지
print(a / b) # 몫
print(a ** b) # 제곱
```

### operation priorities (연산자 우선순위)
- 기본적인 수학의 연산자와 동일
- 강제로 연산을 선수하기 위해선, 괄호()를 사용

```python
a = 10
b = 5

print(a + b * 4) # 곱하기가 우선  10 + (5 * 4)
print((a + b) * 4) # 우선순위 변경 시
```

- 연습문제) 다음의 각 a의 출력값은?

```python
a = 9
print(a)
print(a - 3)
print(a)
```

### expression evaluation & assignment(식평가 & 대입)
- 변수의 값이 변경되기 위해서는 =를 사용하여 대입이 발생하는 경우에만 해당

```python
a = 9
print(a)
a = a - 3
print(a)
```
